```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_json(r"C:\Users\Lenovo\Documents\Web scraping\IMDB\oneplus_mobiles.json")
df
```

```python
df['Price'] = df['Price'].str.replace('₹','')
df['Price'] = df['Price'].str.replace(',','')
df['Price'] = df['Price'].astype(float)
```

```python
df['Battery'] = df['Battery'].str.replace(' mAh Battery', '').astype(float)
```

```python
df
```

```python
df.describe()
```

### Rating:
   - The average rating is approximately 4.27, with a minimum rating of 3.60 and a maximum of 4.80
   - The majority of ratings fall between 4.20 (25th percentile) and 4.40 (75th percentile).
   
### Price:
   - The average price is around 35,544.38, ranging from a minimum of 12,698 to a maximum of 140,000
   - Prices are relatively spread out, as indicated by a standard deviation of approximately 18,755.75

```python
sns.lineplot(x=df['Rating'],y=df['Price'])
plt.title('Rating vs Price')
plt.xlabel('Rating')
plt.ylabel('Price')
```

#### Rating vs Price
   - Phones with lower ratings (between 3.6 and 4.0) are more budget-friendly
   - Phones with mid-range ratings (around 4.2) strike a balance between price and quality.
   - Phones with high ratings (above 4.6) come with a higher price tag, indicating superior 
        quality or additional features

```python
plt.hist(df['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Price Distribution of OnePlus Mobiles')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

#### Price Distribution of OnePlus Mobiles
  - Most OnePlus mobiles fall within the price range of 20,000 to 40,000
  - There are several models priced between 40,000 and 60,000
  - A few premium models are priced around 120,000 to 140,000
  
-  OnePlus offers a variety of options across different price points, catering to both budget-conscious buyers and those seeking premium features

```python
sns.lineplot(x=df['Battery'],y=df['Price'])
plt.title('How Battery Capacity Influences OnePlus Phone Prices')
plt.xlabel('Battery')
plt.ylabel('Price')
plt.show()
```

###### Phones with a battery capacity around 4250 mAh and 5000 mAh are priced higher, indicating these may be premium models or have additional features driving up the cost

###### The decline in price post-5000 mAh suggests that either these models offer less value or there are other factors influencing pricing at this point

```python

camera_counts = df['Camera'].value_counts()

plt.figure(figsize=(10, 10))
plt.pie(camera_counts, labels=camera_counts.index, autopct='%1.1f%%')
plt.title('Distribution of OnePlus Camera Configurations')
plt.show()
```

##### A large segment (39.5%) consists solely of models with a 50MP rear camera
##### The 64MP and 48MP rear camera configurations are also notable, comprising 17.8% and 11.8% respectively
##### Models with a combination of cameras like 48MP + 8MP +2 MP +16 MP and 48 MP +12 MP +16 MP front cameras have smaller segments ranging from 1.7% to 3.4%
